### Combine Labels

```python
padchest.columns, chexpert.columns, chr14.columns, bimcv.columns
```

```python
shared_cols = ['age','sex','frontal/lateral','AP/PA','label','dataset','dataset_ID']
padchest_cols =  ['age','PatientSex_DICOM','Projection','capstone_label','dataset','ImageID']
chexpert_cols = ['Age','Sex','Frontal/Lateral', 'AP/PA' ,'capstone_label','dataset','Rel_path']
chr14_cols = ['Patient Age','Patient Gender','View Position','capstone_label','dataset','Image Index']
bimcv_cols = ['age','gender','loc2','capstone_label','dataset','image_name']
```

```python
padchest_shared = padchest[padchest_cols].copy()
chexpert_shared = chexpert[chexpert_cols].copy()
chr14_shared = chr14[chr14_cols].copy()
bimcv_shared = bimcv[bimcv_cols].copy()
```

```python
def padchest_find_fl(x):
  if x in ['PA','AP_horizontal','AP']:
    return 'Frontal'
  elif x in ['L']:
    return 'Lateral'
  return 'Other'

def padchest_find_ap(x):
  if x in ['PA']:
    return 'PA'
  elif x in ['AP_horizontal','AP']:
    return 'AP'
  return 'N/A'
padchest_shared['frontal/lateral'] = padchest_shared['Projection'].apply(padchest_find_fl)
padchest_shared['AP/PA'] = padchest_shared['Projection'].apply(padchest_find_ap)
padchest_shared.drop(columns=['Projection'], inplace=True)
padchest_shared.columns = ['age','sex','label','dataset','dataset_ID','frontal/lateral','AP/PA']
```

```python
chexpert_shared['Sex'] = chexpert_shared['Sex'].apply(lambda x: 'M' if x == 'Male' else 'F')
chexpert_shared.loc[chexpert_shared['AP/PA'].isna(),'AP/PA'] = 'N/A'
chexpert_shared.loc[chexpert_shared['Frontal/Lateral'].isna(),'Frontal/Lateral'] = 'Other'
chexpert_shared.columns = shared_cols
```

```python
chr14_shared['frontal/lateral'] = 'Frontal'
chr14_shared.columns = ['age','sex','AP/PA','label','dataset','dataset_ID','frontal/lateral']
```

```python
def bimcv_find_fl(x):
  if x in ['vp-pa','vp-ap']:
    return 'Frontal'
  elif x in ['vp-ll','vp-lateral']:
    return 'Lateral'
  return 'Other'

def bimcv_find_ap(x):
  if x in ['vp-pa']:
    return 'PA'
  elif x in ['vp-ap']:
    return 'AP'
  return 'N/A'
bimcv_shared['frontal/lateral'] = bimcv_shared['loc2'].apply(bimcv_find_fl)
bimcv_shared['AP/PA'] = bimcv_shared['loc2'].apply(bimcv_find_ap)
bimcv_shared.drop(columns=['loc2'], inplace=True)
bimcv_shared.columns = ['age','sex','label','dataset','dataset_ID','frontal/lateral','AP/PA']
```

```python
all_shared = pd.concat([padchest_shared, chexpert_shared, chr14_shared, bimcv_shared])
all_shared.reset_index(drop=True, inplace=True)
```

```python
all_shared['filename'] = all_shared.index.astype(str) + '_' + all_shared['dataset'] + '.png'
```

```python
all_shared.to_csv('/content/gcp_storage/datasets/final_dataset/merged_labels.csv', index=False)
```

## Copy

```python
WORK_DIR = '/content/gcp_storage/datasets/final_dataset'

label0_dir = 'neg'
label1_dir = 'pos_non_COVID'
label2_dir = 'pos_COVID'

label0_DIR = os.path.join(WORK_DIR, label0_dir)
label1_DIR = os.path.join(WORK_DIR, label1_dir)
label2_DIR = os.path.join(WORK_DIR, label2_dir)

os.makedirs(label0_DIR, exist_ok=True)
os.makedirs(label1_DIR, exist_ok=True)
os.makedirs(label2_DIR, exist_ok=True)
```

```python
def merge_dataset(df, src_dir,dst_dir):
  print(len(df))
  for i in range(len(df)):
    line = df.iloc[i]
    src = os.path.join(src_dir, line['dataset_ID'])
    dst = os.path.join(dst_dir, line['filename'])
    shutil.copyfile(src, dst)
```
