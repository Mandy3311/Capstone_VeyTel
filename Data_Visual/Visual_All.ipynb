```python
data = all_shared[all_shared['sex']!='O']['sex'].value_counts()

plt.figure(figsize=(8, 8))

def func(pct, allvals):
    absolute = int(pct/100.*sum(allvals))
    for i, val in enumerate(allvals):
      if abs(absolute - val) < 2:
          label = allvals.index[i]
          break
    return f"{label}\n{pct:.1f}%\n({absolute:d})"

plt.pie(data, startangle=140,radius=1.0,autopct=lambda pct: func(pct, data),pctdistance=0.85,
       wedgeprops=dict(width=0.3, edgecolor='w'))
plt.title('Distribution of Sex')
plt.show()
```

```python
data = all_shared['frontal/lateral'].value_counts()
appa = all_shared['AP/PA'].fillna('N/A').value_counts()
plt.figure(figsize=(8, 8))

def func(pct, allvals):
    absolute = int(pct/100.*sum(allvals))
    for i, val in enumerate(allvals):
      if abs(absolute - val) < 2:
          label = allvals.index[i]
          break
    return f"{label}\n{pct:.1f}%\n({absolute:d})"

plt.pie(data, startangle=140,radius=1.0,autopct=lambda pct: func(pct, data),pctdistance=0.85,
       wedgeprops=dict(width=0.3, edgecolor='w'))
plt.pie(appa,radius=0.7, startangle=140,autopct=lambda pct: func(pct, appa),pctdistance=0.75,colors = ['lightblue','lightcyan','white'],
       wedgeprops=dict(width=0.3, edgecolor='w'))

plt.title('Distribution of Frontal/Lateral Types')
plt.show()
```
